```python
def displayW():
    """
    Returns a list of '*' combinations forming W on grid
    """
    w = ["*     *",
         "*     *",
         "*  *  *",
         "* * * *",
         "*     *"]
    return w

def displayO():
    """
    Returns a list of '*' combinations forming O on grid
    """
    o = ["  * * * ",
         "*       *",
         "*       *",
         "*       *",
         "  * * * "]
    return o

def displayR():
    """
    Returns a list of '*' combinations forming R on grid
    """
    R = ["* * * * ",
         "*      *",
         "* * * * ",
         "*   *",
         "*     *"]
    return R

def displayL():
    """
    Returns a list of '*' combinations forming L on grid
    """
    L = ["*",
         "*",
         "*",
         "*",
         "* * * * *"]
    return L

def displayD():
    """
    Returns a list of '*' combinations forming D on grid
    """
    D = ["* * * * ",
         "*       *",
         "*       *",
         "*       *",
         "* * * * "]
    return D

def displayE():
    """
    Returns a list of '*' combinations forming E on grid
    """
    E = ["* * * * *",
         "*",
         "* * * * *",
         "*",
         "* * * * *"]
    return E

def display_letters(letter):
    """
    Calls respective functions for each letter on grid
    """
    if letter == 'W':
        return displayW()
    elif letter == 'O':
        return displayO()
    elif letter == 'R':
        return displayR()
    elif letter == 'L':
        return displayL()
    elif letter == 'D':
        return displayD()
    elif letter == 'E':
        return displayE()

def stringDisplay(word):
    """
    Calls display_letter function with individual letters 
    from the user-input word
    """
    letters = [display_letters(letter) for letter in word]
    max_width = 9  # maximum width of a letter grid
    for i in range(5):
        for letter in letters:
            print(letter[i].ljust(max_width), end="  ")  # pad each letter to max width i.e. left justified with max_width 9 to avoid overlapping
        print()

def main():
    """
    Main function to read input from the user and call other functions
    """
    while True:
        inputWord = input("Enter a word with letters 'W,O,R,L,D,E' to generate grid or 'END' to stop: ").strip()
        if inputWord == 'END':  # END is case sensitive
            break
        elif all(ch in 'WORLDE' for ch in inputWord.upper()):
            stringDisplay(inputWord.upper())
        else:
            print("Invalid word. Try a word using only the specified letters.")

if __name__ == "__main__":
    main()
```
